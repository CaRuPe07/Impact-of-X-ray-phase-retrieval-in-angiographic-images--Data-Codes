```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 15))
ax[0].imshow(PhasePCN[:,X1:X2], cmap="bone")
ax[1].imshow(PhasePCN[:,X5:X6], cmap="bone")
fig.subplots_adjust(wspace=0.1, hspace=0.1,top=0.9, bottom=0.05, left=0, right=1)
plt.show()
```

```python
meanB=np.mean(np.concatenate((RAWCN[:,X1:X2],RAWCN[:,X5:X6]),1))
meanS=np.mean(RAWCN[Y1:Y2,X3:X4])
destB=np.std(np.concatenate((RAWCN[:,X1:X2],RAWCN[:,X5:X6]),1))
destS=np.std(RAWCN[Y1:Y2,X3:X4])**2)
CNR=(np.abs(meanB-meanS))/destB
print("Attenuation image/The meanB is",np.round(meanB,3))
print("Attenuation image/The meanS is",np.round(meanS,3))
print("Attenuation image/The desB is",np.round(destB,3))
print("Attenuation image/The desS is",np.round(destS,3))
print("Attenuation image/The CNR is", np.round(CNR,3))
```

```python
meanBP=np.mean(np.concatenate((PhasePCN[:,X1:X2],PhasePCN[:,X5:X6]),1))
meanSP=np.mean(PhasePCN[Y1:Y2,X3:X4])
destBP=np.std(np.concatenate((PhasePCN[:,X1:X2],PhasePCN[:,X5:X6]),1))
destSP=np.std(PhasePCN[Y1:Y2,X3:X4])
CNRP=(np.abs(meanBP-meanSP))/destBP
print("Phase image/The meanB is",np.round(meanBP,3))
print("Phase image/The meanS is",np.round(meanSP,3))
print("Phase image/The desB is",np.round(destBP,3))
print("Phase image/The desS is",np.round(destSP,3))
print("Phase image/The CNR is", np.round(CNRP,3))
```
