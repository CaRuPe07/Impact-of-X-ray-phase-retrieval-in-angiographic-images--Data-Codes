```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.ticker as ticker
import scipy.stats
```

```python
def func(x, A, m):
    k = A*x+m
    return k
```

```python
IDim=[0.92,1.85,2.82]
NEA=[0.99,0.94,0.90]
NSA=[1.02,0.97,0.94]
NES=[1.16,1.28,1.38]
NSS=[1.12,1.19,1.25]
NEI=[1.28,1.47,1.64]
NSI=[1.17,1.43,1.67]
CEA=[0.92,0.87,0.85]
CSA=[0.85,0.80,0.78]
CES=[0.87,1.20,1.42]
CSS=[0.93,1.07,1.24]
CEI=[1.05,1.39,1.72]
CSI=[1.04,1.31,1.53]
```

```python
resNEA, covNEA = curve_fit(func, IDim, NEA)
resNSA, covNSA = curve_fit(func, IDim, NSA)
resNES, covNES = curve_fit(func, IDim, NES)
resNSS, covNSS = curve_fit(func, IDim, NSS)
resNEI, covNEI = curve_fit(func, IDim, NEI)
resNSI, covNSI = curve_fit(func, IDim, NSI)
resCEA, covCEA = curve_fit(func, IDim, CEA)
resCSA, covCSA = curve_fit(func, IDim, CSA)
resCES, covCES = curve_fit(func, IDim, CES)
resCSS, covCSS = curve_fit(func, IDim, CSS)
resCEI, covCEI = curve_fit(func, IDim, CEI)
resCSI, covCSI = curve_fit(func, IDim, CSI)
```

```python
NEAn=[]
NSAn=[]
NESn=[]
NSSn=[]
NEIn=[]
NSIn=[]
CEAn=[]
CSAn=[]
CESn=[]
CSSn=[]
CEIn=[]
CSIn=[]
for i in range(len(IDim)):
    NEAn.append(func(IDim[i],resNEA[0],resNEA[1]))
    NSAn.append(func(IDim[i],resNSA[0],resNSA[1]))
    NESn.append(func(IDim[i],resNES[0],resNES[1]))
    NSSn.append(func(IDim[i],resNSS[0],resNSS[1]))
    NEIn.append(func(IDim[i],resNEI[0],resNEI[1]))
    NSIn.append(func(IDim[i],resNSI[0],resNSI[1]))
    CEAn.append(func(IDim[i],resCEA[0],resCEA[1]))
    CSAn.append(func(IDim[i],resCSA[0],resCSA[1]))
    CESn.append(func(IDim[i],resCES[0],resCES[1]))
    CSSn.append(func(IDim[i],resCSS[0],resCSS[1]))
    CEIn.append(func(IDim[i],resCEI[0],resCEI[1]))
    CSIn.append(func(IDim[i],resCSI[0],resCSI[1]))
```

```python
IDDim=[0.7,0.92,1.2,1.5,1.85,2.82]
NSSSA=[1.040,1.028,0.997,0.989,0.982,0.943]
NSSSI=[1.214,1.263,1.309,1.355,1.409,1.534]
NSSSS=[1.106,1.128,1.146,1.170,1.195,1.258]
CSSSA=[0.772,0.787,0.763,0.805,0.814,0.842]
CSSSI=[0.975,1.026,1.106,1.161,1.304,1.500]
CSSSS=[0.848,0.849,0.927,1.009,1.128,1.197]
```

```python
resNSSSA, covNSSSA = curve_fit(func, IDDim, NSSSA)
resNSSSS, covNSSSS = curve_fit(func, IDDim, NSSSS)
resNSSSI, covNSSSI = curve_fit(func, IDDim, NSSSI)
resCSSSA, covCSSSA = curve_fit(func, IDDim, CSSSA)
resCSSSS, covCSSSS = curve_fit(func, IDDim, CSSSS)
resCSSSI, covCSSSI = curve_fit(func, IDDim, CSSSI)
```

```python
resNSSSS
```

```python
resCSSSS
```

```python
NSSSAn=[]
NSSSSn=[]
NSSSIn=[]
CSSSAn=[]
CSSSSn=[]
CSSSIn=[]
for i in range(len(IDDim)):
    NSSSAn.append(func(IDDim[i],resNSSSA[0],resNSSSA[1]))
    NSSSSn.append(func(IDDim[i],resNSSSS[0],resNSSSS[1]))
    NSSSIn.append(func(IDDim[i],resNSSSI[0],resNSSSI[1]))
    CSSSAn.append(func(IDDim[i],resCSSSA[0],resCSSSA[1]))
    CSSSSn.append(func(IDDim[i],resCSSSS[0],resCSSSS[1]))
    CSSSIn.append(func(IDDim[i],resCSSSI[0],resCSSSI[1]))
```

```python
residualsNSSSA = np.array(NSSSA)-np.array(NSSSAn)
ss_resNSSSA = np.sum(residualsNSSSA**2)
residualsNSSSS = np.array(NSSSS)-np.array(NSSSSn)
ss_resNSSSS = np.sum(residualsNSSSS**2)
residualsNSSSI = np.array(NSSSI)-np.array(NSSSIn)
ss_resNSSSI = np.sum(residualsNSSSI**2)
residualsCSSSA = np.array(CSSSA)-np.array(CSSSAn)
ss_resCSSSA = np.sum(residualsCSSSA**2)
residualsCSSSS = np.array(CSSSS)-np.array(CSSSSn)
ss_resCSSSS = np.sum(residualsCSSSS**2)
residualsCSSSI = np.array(CSSSI)-np.array(CSSSIn)
ss_resCSSSI = np.sum(residualsCSSSI**2)

ss_totNSSSA = np.sum((NSSSA-np.mean(NSSSA))**2)
ss_totNSSSS = np.sum((NSSSS-np.mean(NSSSS))**2)
ss_totNSSSI = np.sum((NSSSI-np.mean(NSSSI))**2)
ss_totCSSSA = np.sum((CSSSA-np.mean(CSSSA))**2)
ss_totCSSSS = np.sum((CSSSS-np.mean(CSSSS))**2)
ss_totCSSSI = np.sum((CSSSI-np.mean(CSSSI))**2)

RcNSSSA=1 - (ss_resNSSSA / ss_totNSSSA)
RcNSSSS=1 - (ss_resNSSSS / ss_totNSSSS)
RcNSSSI=1 - (ss_resNSSSI / ss_totNSSSI)
RcCSSSA=1 - (ss_resCSSSA / ss_totCSSSA)
RcCSSSS=1 - (ss_resCSSSS / ss_totCSSSS)
RcCSSSI=1 - (ss_resCSSSI / ss_totCSSSI)
print("The correlation coeficient for NAir is",np.round(RcNSSSA,3))
print("The correlation coeficient for NBlood is",np.round(RcNSSSS,3))
print("The correlation coeficient for NIodine is",np.round(RcNSSSI,3))
print("The correlation coeficient for CAir is",np.round(RcCSSSA,3))
print("The correlation coeficient for CBlood is",np.round(RcCSSSS,3))
print("The correlation coeficient for CIodine is",np.round(RcCSSSI,3))
```

```python
plt.plot(IDDim, NSSSA, 'o', label="MC Simulations-Air", color="cyan")
plt.plot(IDDim, NSSSAn, '--', color="cyan")
plt.plot(IDDim, NSSSS, 'o', label="MC Simulations-A.blood", color="darkred")
plt.plot(IDDim, NSSSSn, '--', color="darkred")
plt.plot(IDDim, NSSSI, 'o', label="MC Simulations-P.iodine", color="lime")
plt.plot(IDDim, NSSSIn, '--', color="lime")
plt.xlabel("Diameter (mm)",fontsize=14)
plt.ylabel(r'$ N=\frac{CNR_{PI}}{CNR_{AI}} $',fontsize=14)
plt.xlim(0.6,3)
plt.ylim(0.5,2.1)
plt.grid(linestyle = '--', linewidth = 0.5)
plt.legend(loc="upper left", mode = "expand", ncol = 2,fontsize=10.5)
plt.savefig("NSimulationsMD.jpeg", bbox_inches='tight', transparent=True, pad_inches=0)
```
