```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.ticker as ticker
import scipy.stats
```

```python
def func(x, A, m):
    k = A*x+m
    return k
```

```python
IDim=np.array([0.70,0.92,1.20,1.50,1.85,2.35,2.82])
ErrorDim=np.array([0,0,0,0,0,0,0])
CPCS=np.array([1.050,1.153,1.114,1.235,1.300,1.445,1.485])
CPCI=np.array([1.166,1.230,1.349,1.447,1.577,1.659,1.804])
ErrorCPC=np.array([0.028,0.028,0.028,0.028,0.028,0.028,0.028])
CNRPCS=np.array([1.214,1.229,1.253,1.280,1.302,1.334,1.354])
CNRPCI=np.array([1.329,1.367,1.424,1.465,1.513,1.579,1.634])
ErrorCNRPC=np.array([0.014,0.014,0.014,0.014,0.014,0.014,0.014])
```

```python
resCPCS, covCPCS = curve_fit(func, IDim, CPCS)
resCPCI, covCPCI = curve_fit(func, IDim, CPCI)
resCNRPCS, covCNRPCS = curve_fit(func, IDim, CNRPCS)
resCNRPCI, covCNRPCI = curve_fit(func, IDim, CNRPCI)
yFitCPCS=func(IDim,*resCPCS)
yFitCPCI=func(IDim,*resCPCI)
yFitCNRPCS=func(IDim,*resCNRPCS)
yFitCNRPCI=func(IDim,*resCNRPCI)
```

```python
plt.errorbar(IDim, CPCS, ErrorCPC, ErrorDim, "o",label="MC-A.blood", color="darkred")
plt.errorbar(IDim, CPCI, ErrorCPC, ErrorDim, "^",label="MC-P.iodine", color="blue")
plt.plot(IDim, yFitCPCS, '--', label="Fit-A.blood", color="darkred")
plt.plot(IDim, yFitCPCI, '--', label="Fit-P.iodine", color="blue")
plt.xlabel("Diameter (mm)",fontsize=14)
plt.ylabel(r'$ \frac{C_{PI}}{C_{CI}} $',fontsize=20)
plt.xlim(0.4,3)
plt.ylim(1,2.1)
plt.grid(linestyle = '--', linewidth = 0.5)
plt.text(0.7, 1.7, r'$ \frac{C_{PI}}{C_{CI}}= $', fontsize=19, color='darkred')
plt.text(0.7, 1.9, r'$ \frac{C_{PI}}{C_{CI}}= $', fontsize=19, color='blue')
plt.text(1.02, 1.71, str(round(resCPCS[0], 2))+" D+"+str(round(resCPCS[1], 2)), fontsize=14, color='darkred')
plt.text(1.02, 1.91, str(0.29)+" D+"+str(round(resCPCI[1], 2)), fontsize=14, color='blue')
#plt.text(28, 13.3, "Fs="+str(round(resVM[0], 2))+"V+"+str(round(resVM[1], 2)), fontsize=14, color='black')
#plt.text(28, 13, "$R^{2}$="+str(round(RcVM, 2)), fontsize=14, color='black')
plt.legend(bbox_to_anchor=(0.89, 1.27),mode = "Expand", ncol=2,fontsize=12.5)
#plt.savefig("CSimulationsMD.pdf", bbox_inches='tight', transparent=True, pad_inches=0,dpi=199)
```

```python
plt.errorbar(IDim, CNRPCS, ErrorCNRPC, ErrorDim, "o",label="MC-A.blood", color="darkred")
plt.errorbar(IDim, CNRPCI, ErrorCNRPC, ErrorDim, "^",label="MC-P.iodine", color="blue")
plt.plot(IDim, yFitCNRPCS, '--', label="Fit-A.blood", color="darkred")
plt.plot(IDim, yFitCNRPCI, '--', label="Fit-P.iodine", color="blue")
plt.xlabel("Diameter (mm)",fontsize=14)
plt.ylabel(r'$ \frac{CNR_{PI}}{CNR_{CI}} $',fontsize=20)
plt.xlim(0.4,3)
plt.ylim(1,2.1)
plt.grid(linestyle = '--', linewidth = 0.5)
plt.text(0.7, 1.7, r'$ \frac{CNR_{PI}}{CNR_{CI}}= $', fontsize=19, color='darkred')
plt.text(0.7, 1.9, r'$ \frac{CNR_{PI}}{CNR_{CI}}= $', fontsize=19, color='blue')
plt.text(1.2, 1.71, str(round(resCNRPCS[0], 2))+" D+"+str(round(resCNRPCS[1], 2)), fontsize=14, color='darkred')
plt.text(1.2, 1.91, str(round(resCNRPCI[0], 2))+" D+"+str(round(resCNRPCI[1], 2)), fontsize=14, color='blue')
plt.legend(bbox_to_anchor=(0.89, 1.27),mode = "Expand", ncol=2,fontsize=12.5)
#plt.savefig("CNRSimulationsMD.pdf", bbox_inches='tight', transparent=True, pad_inches=0,dpi=199)
```
